```python
from sympy import *
```

```python
A,B,C,D,y,xi,alpha,beta,mu,k,H = symbols('A B C D y xi alpha beta mu k H')
ux = I *(xi*A*sin(alpha*y) + B*cos(alpha*y) + beta*(C*cos(beta*y)-D*sin(beta*y)))
uy = alpha*(A*cos(alpha*y) - B*sin(alpha*y)) + xi*(C*sin(beta*y) + D*cos(beta*y))
```

```python
BB = solve(ux.subs(y,0),B)[0]
BB
```

```python
AA = solve(uy.subs(y,0),A)[0]
AA
```

```python
tauyy = mu*((2*alpha**2 - k**2*(xi**2+alpha**2))*(A*sin(alpha*y) + B*cos(alpha*y))
           +2*beta*xi*(C*cos(beta*y) - B*sin(alpha*y)))
tauxy = I*mu*( 2*alpha*xi*(A*cos(alpha*y) - B*sin(alpha*y)) - (beta**2 - xi**2)*(C*sin(beta*y) + D*cos(beta*y)))
```

```python
CC = solve(tauyy.subs(y,H).subs(A,AA).subs(B,BB),C)[0]
CC
```

```python
dispersion_relation = tauxy.subs(y,H).subs(A,AA).subs(B,BB).subs(C,CC).subs(D,1).subs(mu,1)
```
